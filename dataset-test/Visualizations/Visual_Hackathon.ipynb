```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
df = pd.read_excel('~/Desktop/dataset.xlsx')
df.head()
```

```python
mydict = {'Male':0,'Female':0}


for item in df['Gender']:
    if item == 'M':
        mydict['Male'] +=1
    else:
        mydict['Female'] +=1

mydict['Total'] = 1426
print(mydict)

```

```python
dictionary = plt.figure()


plt.bar(range(len(mydict)), mydict.values(), align='center')
plt.xticks(range(len(mydict)), mydict.keys())
```
